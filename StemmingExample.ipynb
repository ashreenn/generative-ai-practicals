In this section we will learn what is Stemming. It is a process in text processing where we cut words down to there root form. The root form might not always be a real word in english but it helps in treating similar words as the same.


```python
##classification problem
##comments of products is a positive review or negative review
##reviews----> eating,eat,eaten[going,gone,goes]---->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalize"]
```

PorterStemmer


```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer() #storing the object into a variable
```

```python
for word in words:
  print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem("congratulations") #notice the disadvantage as well
```

**RegexpStemmer Class**:
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer("ing$|s$|e$|able$",min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingeating")  #only removes the last word
```

**Snowball Stemmer**: improved version of the Porter Stemmer that stems words more accurately and supports multiple languages.

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer("english") #We give "english" to tell the Snowball Stemmer which language rules to use.
```

```python
for word in words:
  print(word+"---->"+snowballstemmer.stem(word))
```
