```python
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

nltk.download('stopwords')
```

```python
messages = pd.DataFrame({
    'text': [
        'I love machine learning',
        'Machine learning is amazing',
        'I do not like bad models',
        None
    ]
})
```

```python
ps = PorterStemmer()
corpus = []

for i in range(len(messages)):
    if pd.isna(messages.iloc[i, 0]):
        continue

    review = messages.iloc[i, 0]
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review
              if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
cv1 = CountVectorizer(ngram_range=(1,1))
X1 = cv1.fit_transform(corpus)

print(cv1.get_feature_names_out())
print(X1.toarray())
```

```python
cv2 = CountVectorizer(ngram_range=(2,2))
X2 = cv2.fit_transform(corpus)

print(cv2.get_feature_names_out())
print(X2.toarray())
```
