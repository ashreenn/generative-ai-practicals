Text Preprocessing Stopwords with NLTK

```python
## Speech of DR APJ Abdul Kalam
paragraph="""Dr. A. P. J. Abdul Kalam was a great Indian scientist and the 11th President of India. He is known as the Missile Man of India for his important role in the development of India’s missile and space programs. Born on 15 October 1931 in Rameswaram, Tamil Nadu, he rose from humble beginnings to become a respected national leader. He served as President from 2002 to 2007 and was loved by students for his inspiring thoughts and simple lifestyle. Dr. Kalam believed strongly in education, hard work, and dreaming big, and his life continues to motivate millions of young people in India and around the world.
"""
```

```python
import nltk
nltk.download('punkt')

nltk.download('punkt_tab')

```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english') #we can check for different languages as well
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
##apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=' '.join(words) #converting all the list of words into sentences
```
