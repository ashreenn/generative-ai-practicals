**Wordnet Lemmatizer**: this technique is like stemming. The output we will get after lemmatization is called lemma which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same things.
NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus.This class uses morphy() functin to the WordNet CorpusReader class to find a lemma.Let us understand it with an example-

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
## POS	Meaning
##'n'	noun
##'v'	verb
##'a'	adjective
##'r'	adverb

##pos stands for part of speech.
##It tells the lemmatizer what kind of word it is so it can choose the correct base form.

lemmatizer.lemmatize("going",pos="a")
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally"]
```

```python
for word in words:
  print(word+"---->"+lemmatizer.lemmatize(word))
```
