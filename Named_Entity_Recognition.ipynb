```python
sentence="The Eiffel Tower was built from 1887 to 1890 by French engineer Gustave Eiffel,whose company specialized metal frameworks and structures."
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tag_elements)
```
