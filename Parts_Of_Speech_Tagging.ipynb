```python
## we will find the Pos Tag

##importing imp resources
import nltk
nltk.download('stopwords')
import nltk
nltk.download('averaged_perceptron_tagger_eng')


from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

stemmer=PorterStemmer()
sentences=nltk.sent_tokenize(paragraph)
```

In the following code segment we are looping thriugh each sentence in sentences list using index i. Breaks the current sentence into words(tokens). Then removing stopwords and applying pos to the remaining words.

```python
for i in range(len(sentences)):
  words=nltk.word_tokenize(sentences[i])
  words=[word for word in words if word not in set(stopwords.words('english'))]
  poss_tag=nltk.pos_tag(words)
  print(poss_tag)
```
